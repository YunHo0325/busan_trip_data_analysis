# 데이터 전처리

```python
def li2idx(li):
    tmp = [i.split(', ') for i in li]
    tmp = [", ".join(i).split(',') for i in tmp]
    group_set, tmp2 = [], []
    for i in tmp:
        tmp2.extend(i)
    for i in tmp2:
        group_set.append(i.strip())
    group_set = list(set(group_set))
    
    group_dic = dict()
    for i in group_set:
        group_dic[i] = [0]*len(li)

    for idx, i in enumerate(li):
        for group in group_set:

            if group in i:
                group_dic[group][idx] = 1
                
    return group_dic
```

```python
li = list(df_custom['GRP_NM'])

result_dic = li2idx(list(df_custom['GRP_NM']))
for k, i in result_dic.items():
    df_custom[k] = i
```

```python
result_dic = li2idx(list(df_custom['SEASON_NM']))
for k, i in result_dic.items():
    df_custom[k] = i

df_custom['사계절'] = df_custom['사게절'] + df_custom['사계절']
df_custom['봄'] = df_custom['봄'] + df_custom['사계절']
df_custom['여름'] = df_custom['여름'] + df_custom['사계절']
df_custom['가을'] = df_custom['가을'] + df_custom['사계절']
df_custom = df_custom.drop(columns=['사게절'])
df_custom.rename(columns={"사계절" : "겨울"}, inplace=True)
```

```python
li = list(df_custom['IEM_NM'])

result_dic = li2idx(list(df_custom['IEM_NM']))
for k, i in result_dic.items():
    df_custom[k] = i
```

```python
result_dic = li2idx(list(df_custom['CL_NM']))
for k, i in result_dic.items():
    df_custom[k] = i
```

```python
df_custom.to_csv('../dataset/custom_preprocessing.csv', index=False)
```
